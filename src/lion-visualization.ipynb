```python
import matplotlib.pyplot as plt
import numpy as np

# Increase text sizes globally
plt.rcParams.update({
    'font.size': 14,          # General text size
    'axes.titlesize': 16,     # Title font size
    'axes.labelsize': 14,     # Label font size
    'xtick.labelsize': 12,    # X-tick label size
    'ytick.labelsize': 12,    # Y-tick label size
    'legend.fontsize': 12     # Legend font size
})

# Data from the table (accuracies for each batch size)
batch_sizes = [64, 128, 256, 512, 1024, 2048]
accuracies = [
    [41.11, 41.95, 41.99, 42.00, 41.72],  # Accuracies for batch size 64
    [42.74, 34.50, 35.79, 36.90, 35.72],  # Accuracies for batch size 128
    [43.35, 43.67, 40.29, 41.62, 41.10],  # Accuracies for batch size 256
    [43.78, 44.46, 41.37, 45.79, 44.08],  # Accuracies for batch size 512
    [41.08, 41.80, 43.85, 39.77, 42.12],  # Accuracies for batch size 1024
    [37.49, 38.20, 34.78, 40.08, 38.02]   # Accuracies for batch size 2048
]

# Sort batch sizes and corresponding accuracies in the order 64, 128, 256, 512, 1024, 2048
sorted_batch_sizes = [64, 128, 256, 512, 1024, 2048]

# Create a sorted list of accuracies to match the sorted batch sizes
sorted_accuracies = [
    np.mean(accuracies[i]) for i in [0, 1, 2, 3, 4, 5]  # Reorder the accuracies accordingly
]

# Calculate the standard deviation for each batch size in the sorted order
sorted_std_devs = [
    np.std(accuracies[i]) for i in [0, 1, 2, 3, 4, 5]  # Reorder the standard deviations accordingly
]

# Plotting
plt.figure(figsize=(10, 6))

# Plot the average accuracies connected by a line
plt.plot(sorted_batch_sizes, sorted_accuracies, marker='o', color='b', label='Average Accuracy')

# Fill the area between (average_accuracy - std_dev) and (average_accuracy + std_dev)
plt.fill_between(sorted_batch_sizes, 
                 np.array(sorted_accuracies) - np.array(sorted_std_devs), 
                 np.array(sorted_accuracies) + np.array(sorted_std_devs), 
                 color='blue', alpha=0.2, label='Standard Deviation Spread')

# Plot error bars (standard deviation) around the points
plt.errorbar(sorted_batch_sizes, sorted_accuracies, yerr=sorted_std_devs, fmt='o', color='b', 
             ecolor='red', capsize=5, label='Error Bars (Std Dev)')

# Explicitly label the x-axis and y-axis
plt.xlabel('Batch Size')
plt.ylabel('Average Accuracy (%)')

# Title of the plot
plt.title('Batch Size vs Average Accuracy')

# Use logarithmic scale for batch size
plt.xscale('log')

# Explicitly show the batch sizes on the x-axis (log scale will display them as powers of 10)
plt.xticks(sorted_batch_sizes, labels=[str(bs) for bs in sorted_batch_sizes])

# Grid and legend
plt.grid(True)
plt.legend()

# Save the plot to a high-resolution PNG file
save_path = 'batch_size_vs_accuracy.png'
plt.savefig(save_path, dpi=300)  # Save with high resolution

# Show the plot
plt.tight_layout()
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Increase text sizes globally
plt.rcParams.update({
    'font.size': 14,          # General text size
    'axes.titlesize': 16,     # Title font size
    'axes.labelsize': 14,     # Label font size
    'xtick.labelsize': 12,    # X-tick label size
    'ytick.labelsize': 12,    # Y-tick label size
    'legend.fontsize': 12     # Legend font size
})

# Data for the accuracies at different learning rates
accuracies = [
    [32.36, 32.72, 33.79, 34.33, 29.98],  # Accuracy for learning rate 0.1
    [40.99, 40.52, 40.62, 42.58, 40.97],  # Accuracy for learning rate 0.01
    [44.92, 45.21, 45.13, 45.09, 44.24],  # Accuracy for learning rate 0.001
    [40.88, 40.66, 41.77, 40.52, 40.76]   # Accuracy for learning rate 0.0001
]

# Learning rates corresponding to the above accuracies
learning_rates = [0.1, 0.01, 0.001, 0.0001]

# Calculate the average accuracy and standard deviation for each learning rate
average_accuracies = [np.mean(acc) for acc in accuracies]
std_devs = [np.std(acc) for acc in accuracies]

# Plotting
plt.figure(figsize=(10, 6))

# Plot the average accuracies connected by a line
plt.plot(learning_rates, average_accuracies, marker='o', color='b', label='Average Accuracy')

# Fill the area between (average_accuracy - std_dev) and (average_accuracy + std_dev)
plt.fill_between(learning_rates, 
                 np.array(average_accuracies) - np.array(std_devs), 
                 np.array(average_accuracies) + np.array(std_devs), 
                 color='blue', alpha=0.2, label='Standard Deviation Spread')

# Plot error bars (standard deviation) around the points
plt.errorbar(learning_rates, average_accuracies, yerr=std_devs, fmt='o', color='b', 
             ecolor='red', capsize=5, label='Error Bars (Std Dev)')

# Explicitly label the x-axis and y-axis
plt.xlabel('Learning Rate')
plt.ylabel('Average Accuracy (%)')

# Title of the plot
plt.title('Learning Rate vs Average Accuracy')

# Set log scale for x-axis and flip the axis (reverse x-axis)
plt.xscale('log')
plt.gca().invert_xaxis()  # Flip the x-axis

# Explicitly show the learning rates on the x-axis
plt.xticks(learning_rates, labels=[str(lr) for lr in learning_rates])

# Grid and legend
plt.grid(True)
plt.legend()

# Save the plot to a high-resolution PNG file
save_path = 'learning_rate_vs_accuracy.png'
plt.savefig(save_path, dpi=300)  # Save with high resolution

# Show the plot
plt.tight_layout()
plt.show()
```
